# EDA SET-2

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import trim_mean

data = pd.read_csv("state.csv")

print("Type: ", type(data), "\n\n")
print("Head: ", data.head(10), "\n\n")
print("Tail: ", data.tail(10), "\n\n")

data["Population in Millions"] = data["Population"]/1000000
print(data.head())

data.rename(columns = {'Murder.Rate':'MurderRate'}, inplace = True)
list(data)
```

VISUALIZING POPULATION PER MILLION:

```python
fig, x1 = plt.subplots()
fig.set_size_inches(15,9)

x1 = sns.barplot(x='State', y='Population',
                data = data.sort_values('MurderRate'),
                palette='Set2')

x1.set(xlabel = "State", ylabel = "Population in Millions")
x1.set_title("Population in Millions by State", size = 20)

plt.xticks(rotation =-90)
```

VISUALIZING MURDER RATE PER LAKHS:

```python
fig, x2 = plt.subplots()
fig.set_size_inches(15,  9)
  
  
x2 = sns.barplot(
    x ="State", y ="MurderRate", 
    data = data.sort_values('MurderRate', ascending = 1), 
                                         palette ="husl")
  
x2.set(xlabel ='States', ylabel ='Murder Rate per 100000')
x2.set_title('Murder Rate by State', size = 20)
  
plt.xticks(rotation =-90)
```

STANDARD DAVIATION:

```python
Population_std = data.Population.std()
print("Population std: ", Population_std)

MurderRate_std = data.MurderRate.std()
print("MurderRate std: ", MurderRate_std)
```

vARIANCE:

```python
Population_var = data.Population.var()
print("Population var: ", Population_var)

MurderRate_var = data.MurderRate.var()
print("MurderRate var: ", MurderRate_var)
```

INTER QUARTILE RANGE:

```python
Population_IQR = data.Population.describe()['75%'] - data.Population.describe()['25%']
print("Population IQR: ", Population_IQR)

MurderRate_IQR = data.MurderRate.describe()['75%'] - data.MurderRate.describe()['25%']
print("MurderRate IQR: ", MurderRate_IQR)
```

MEAN ABSOULTE DEVIATION (MAD)

```python
Population_mad = data.Population.mad()
print("Population MAD: ", Population_mad)

MurderRate_mad = data.MurderRate.mad()
print("MurderRate MAD: ", MurderRate_mad)
```
