# 2) Histogram:

Histograms are bar chatrs where y-axis contains the frequency of the variables.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
```


Histtype = 'step':


```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='red',histtype='step')
```


Mutiple data samples in a histogram


```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,color=['green','blue'])
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no. of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,color=['green','blue'],
         label=['Men','Women'])
plt.legend()
```


Horizontal orientation:


```python


plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')

```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no. of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,color=['green','blue'],
         label=['Men','Women'],orientation='horizontal')
plt.legend()
```

# 3) Pie-Chart:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```


Show percentages for every pie. Specify radius to increase chart size:


```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%', shadow=True)
plt.show()
```

Explode:

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0,0,0,0.5])
plt.show()
```


Counterclock and angle properties:


```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0,0,0,0.5],
       startangle=30)
plt.show()
```
