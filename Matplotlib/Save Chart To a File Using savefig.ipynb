# Save Chart To a File Using savefig:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%',radius=1.5,explode=[0,0,0,0,0.5],
        counterclock=True, startangle=30)
plt.savefig("savefig1.jpg",bbox_inches='tight',pad_inches=2)
```
